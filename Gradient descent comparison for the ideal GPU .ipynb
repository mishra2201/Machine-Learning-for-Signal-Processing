### Why is SGD less computationally efficient than GD?

Stochastic radient descnet (SGD) is computationally more efficient than gradient descent. </br> Say we have 100 traning samples of a dataset spanning 10 parameters. Assume 50 epoch. Gradient descent (GD) will calculate the gradient of the loss function with resppect to all 10 parameters of all 100 datasets in parallel and then provide a single update to the model. It will do this 50  times. <br> <b>Total # evalutations = 100 * 10*50 = 50,000.  </b>   </br> </br> 

SGD, say with a batch size of 2, will span all 10 params of 2 datasets (randomly sampled) and provide an update to the model. It will do this 100 / 2 = 50 times per epoch. <br> <b> Total # computations 50*2*10*50 = <b> 50,000</b> times.</b> <br> Though SDG dataset is smaller, the number of operations is still the same. So where does the copmutational overhead come from?


Differentiator here is the way GPUs function. If we had an infinitely parallel GPU, the two approaches would take similar time. Thus GPUs are limited by their finite cache, bandwith, memorry and core contraints.
