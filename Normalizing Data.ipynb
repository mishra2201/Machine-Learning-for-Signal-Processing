### Why normalize data?  </br>

where the elements are vectorized </br></br>
Say for two independent (maybe even dependent?) samples with position vectors x1, x2, </br>
x1 $ \in $ [0,1] </br>
x2 $ \in $ [100, 8000] </br>

----
The role of a NN is to train the neurons to identify certain patterns in data. This "traning" is represented in the form of the weights and biases the model learns. If the training data has wild fluctuations in the input vector, then either we need many epochs of traning to average out the fluctuations or we face uneven gradients. The model may even try to ignore the inputs with smaller values.  </br>

This normalization is often scaled such that the input data is 1 standard deviation. 

### Is scaling across datasets or across features?

##### Apparently, this is known as per-sample vs per feature scaling

My intuition was that per-feature scaling would lead to a loss of information within the sample. This is because the information on the relative intensities between features in a given sample will be lost, say in a time-series data. 

For the model to predict what $t_n$ will look like, it needs to see how $ \sum_0^m t$ affects $t_{m+1}$ so it will train on each $t_m$ across datasets.
